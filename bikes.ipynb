```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
plt.xkcd()
import seaborn as sns
import contextily as cx
```

### Bikes
[Get data here](https://www.opendata.dk/city-of-copenhagen/cykelparkeringstaelling#resource-cykelparkeringstaellinger.csv)

```python
bikes = gpd.read_file('cykeltaellinger_2017.geojson')
```

```python
bikes.head()
```

```python
f, ax = plt.subplots(figsize=(12, 12))

basemap = cx.providers.CartoDB.DarkMatterNoLabels
bikes.plot(ax=ax, alpha=.3, markersize=2)
cx.add_basemap(ax=ax, crs=bikes.crs, source=basemap)
ax.set_axis_off()
```

### Individual Facebook users
[Get data here](https://dataforgood.facebook.com/dfg/tools/high-resolution-population-density-maps)

```python
fb_df = pd.read_csv('dnk_general_2020.csv')
```

```python
fb = gpd.GeoDataFrame(fb_df, geometry=gpd.points_from_xy(fb_df.longitude, fb_df.latitude))
```

```python
f, ax = plt.subplots(figsize=(12,12))

fb.sample(1000000).plot(ax=ax, alpha=.2, markersize=2)
```
